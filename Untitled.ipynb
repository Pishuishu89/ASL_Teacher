```python
# train_asl_model_v3.py

import os
import glob
import cv2
import pickle
import numpy as np
import mediapipe as mp
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf

def landmark_angles(landmarks):
    """
    Compute joint angles for each finger from world landmarks.
    Returns a (10,) array of angles in degrees.
    """
    fingers = {
        'thumb':   [(1,2,3),(2,3,4)],
        'index':   [(5,6,7),(6,7,8)],
        'middle':  [(9,10,11),(10,11,12)],
        'ring':    [(13,14,15),(14,15,16)],
        'pinky':   [(17,18,19),(18,19,20)],
    }
    coords = np.array([[p.x, p.y, p.z] for p in landmarks], dtype=np.float32)
    angles = []
    for joints in fingers.values():
        for a, b, c in joints:
            v1 = coords[a] - coords[b]
            v2 = coords[c] - coords[b]
            cosang = np.dot(v1, v2) / (np.linalg.norm(v1)*np.linalg.norm(v2) + 1e-6)
            angles.append(np.degrees(np.arccos(np.clip(cosang, -1, 1))))
    return np.array(angles, dtype=np.float32)

def tip_distances(landmarks):
    """
    Compute Euclidean distance from each fingertip to the wrist.
    Returns a (5,) array of distances.
    """
    pts = np.array([[p.x, p.y, p.z] for p in landmarks], dtype=np.float32)
    wrist = pts[0]
    tips  = pts[[4,8,12,16,20]]
    return np.linalg.norm(tips - wrist, axis=1)

def hull_area(landmarks, w, h):
    """
    Compute 2D convex-hull area of hand projection.
    landmarks: image-space normalized Landmarks
    w,h: frame width & height
    """
    pts = np.array([[int(p.x*w), int(p.y*h)] for p in landmarks], dtype=np.int32)
    hull = cv2.convexHull(pts)
    return cv2.contourArea(hull)

def main():
    # --- Configuration ---
    DATASET_DIR = r"C:\Users\ishaa\Downloads\SpeakEz\Teacher\asl_alphabet_test"
    IMG_EXTS    = ("*.jpg", "*.png")
    MODEL_PATH  = "asl_letter_model_v3.keras"
    SCALER_PATH = "scaler_v3.pkl"
    LE_PATH     = "label_encoder_v3.pkl"

    # --- MediaPipe setup ---
    hands = mp.solutions.hands.Hands(
        static_image_mode=True,
        model_complexity=1,
        max_num_hands=1,
        min_detection_confidence=0.5
    )

    # --- Data collection (only single-letter folders A–Z) ---
    data, labels = [], []
    for cls in sorted(os.listdir(DATASET_DIR)):
        if len(cls) != 1 or not cls.isalpha():
            continue
        cls_dir = os.path.join(DATASET_DIR, cls)
        for ext in IMG_EXTS:
            for fp in glob.glob(os.path.join(cls_dir, ext)):
                img = cv2.imread(fp)
                if img is None:
                    continue
                h_img, w_img = img.shape[:2]
                rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                res = hands.process(rgb)
                if not res.multi_hand_world_landmarks:
                    continue

                w_lms   = res.multi_hand_world_landmarks[0].landmark
                coords  = np.array([[p.x,p.y,p.z] for p in w_lms], dtype=np.float32).flatten()
                angs    = landmark_angles(w_lms)
                dists   = tip_distances(w_lms)
                img_lms = res.multi_hand_landmarks[0].landmark
                area    = hull_area(img_lms, w_img, h_img)

                feat = np.concatenate([coords, angs, dists, [area]])
                data.append(feat)
                labels.append(cls.upper())

    hands.close()

    X = np.array(data, dtype=np.float32)
    y = np.array(labels)
    print(f">> Loaded {len(X)} samples from classes: {sorted(set(y))}")

    # --- Encode labels & scale features ---
    le    = LabelEncoder().fit(y)
    y_enc = le.transform(y)
    y_ohe = tf.keras.utils.to_categorical(y_enc)

    scaler = StandardScaler().fit(X)
    Xs     = scaler.transform(X)

    # --- Train/validation split ---
    Xtr, Xvl, ytr, yvl = train_test_split(
        Xs, y_ohe,
        test_size=0.2,
        random_state=42,
        stratify=y_ohe
    )

    # --- Build the Keras model ---
    inp_dim = Xtr.shape[1]
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(inp_dim,)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dropout(0.4),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(len(le.classes_), activation='softmax')
    ])

    # --- Ensure all inputs are float32 ---
    Xtr = Xtr.astype(np.float32)
    ytr = ytr.astype(np.float32)
    Xvl = Xvl.astype(np.float32)
    yvl = yvl.astype(np.float32)

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    # --- Train ---
    hist = model.fit(
        Xtr, ytr,
        validation_data=(Xvl, yvl),
        epochs=50,
        batch_size=16,
        callbacks=[
            tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
            tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3)
        ]
    )

    # --- Save model and preprocessors ---
    model.save(MODEL_PATH)
    with open(SCALER_PATH, 'wb') as f:
        pickle.dump(scaler, f)
    with open(LE_PATH, 'wb') as f:
        pickle.dump(le, f)
    print("✅ Main model and preprocessors saved.")

    # --- Plot training history ---
    epochs = range(1, len(hist.history['accuracy']) + 1)
    plt.figure()
    plt.plot(epochs, hist.history['accuracy'], label='train acc')
    plt.plot(epochs, hist.history['val_accuracy'], label='val acc')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    plt.figure()
    plt.plot(epochs, hist.history['loss'], label='train loss')
    plt.plot(epochs, hist.history['val_loss'], label='val loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    # --- Confusion matrix & classification report ---
    y_true = np.argmax(yvl, axis=1)
    y_pred = np.argmax(model.predict(Xvl, verbose=0), axis=1)
    cm     = confusion_matrix(y_true, y_pred)

    fig, ax = plt.subplots(figsize=(10, 10))
    im = ax.imshow(cm, interpolation='nearest', cmap='viridis')
    ax.set_title('Confusion Matrix (counts)')
    fig.colorbar(im, ax=ax)

    tick_marks = np.arange(len(le.classes_))
    ax.set_xticks(tick_marks)
    ax.set_xticklabels(le.classes_, rotation=90)
    ax.set_yticks(tick_marks)
    ax.set_yticklabels(le.classes_)

    thresh = cm.max() / 2
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(
                j, i, str(cm[i, j]),
                ha='center', va='center',
                color='white' if cm[i, j] > thresh else 'black'
            )

    ax.set_ylabel('True label')
    ax.set_xlabel('Predicted label')
    plt.tight_layout()
    plt.show()

    print(classification_report(y_true, y_pred, target_names=le.classes_))

if __name__ == "__main__":
    main()
```

```python
# train_refiners_simple.py

import os, glob, cv2
import numpy as np
import pickle
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# CONFIG
ROOT        = r"C:\Users\ishaa\Downloads\SpeakEz\Teacher\asl_alphabet_test"
IMG_EXTS    = ("*.jpg","*.png")
TARGET_SIZE = (128,128)

CLOSED_CLASSES = ["A","E","O","S","M","N","T"]
BW_CLASSES     = ["B","W"]

def load_images(classes):
    X, y = [], []
    for cls in classes:
        folder = os.path.join(ROOT, cls)
        if not os.path.isdir(folder):
            print(f"⚠️  Missing {folder}, skipping {cls}")
            continue
        for ext in IMG_EXTS:
            for fp in glob.glob(os.path.join(folder, ext)):
                img = cv2.imread(fp)
                if img is None: 
                    continue
                img = cv2.resize(img, TARGET_SIZE)
                X.append(img)
                y.append(cls)
    X = np.array(X, dtype=np.float32) / 255.0
    y = np.array(y)
    print(f"Loaded {len(X)} images for {classes}")
    return X, y

def make_tf_datasets(X, y, batch=32, binary=False):
    # encode labels
    le     = LabelEncoder().fit(y)
    y_enc  = le.transform(y)
    # train/val split
    Xtr,Xvl,ytr,yvl = train_test_split(
        X, y_enc, test_size=0.2, random_state=42, stratify=y_enc
    )
    # build datasets
    train_ds = tf.data.Dataset.from_tensor_slices((Xtr, ytr))
    train_ds = train_ds.shuffle(100).batch(batch).prefetch(1)
    val_ds   = tf.data.Dataset.from_tensor_slices((Xvl, yvl))
    val_ds   = val_ds.batch(batch).prefetch(1)
    return train_ds, val_ds, le

def train_closed_fist():
    X, y = load_images(CLOSED_CLASSES)
    if len(X)==0:
        print("❌ No closed-fist data, skipping")
        return
    train_ds, val_ds, le = make_tf_datasets(X, y, binary=False)

    # clear any tiny cached weights
    cache = os.path.expanduser("~/.keras/models")
    fname = "mobilenet_v2_weights_tf_dim_ordering_tf_kernels_1.0_128_no_top.h5"
    fpath = os.path.join(cache, fname)
    if os.path.exists(fpath) and os.path.getsize(fpath)<5_000_000:
        os.remove(fpath)

    # model
    base = tf.keras.applications.MobileNetV2(
        input_shape=(*TARGET_SIZE,3),
        include_top=False, pooling='avg', weights='imagenet'
    )
    x   = tf.keras.layers.Dropout(0.3)(base.output)
    out = tf.keras.layers.Dense(len(CLOSED_CLASSES), activation='softmax')(x)
    model = tf.keras.Model(base.input, out)
    base.trainable = False

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=10,
        callbacks=[
            tf.keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)
        ]
    )

    model.save("closed_fist_refiner.keras")
    with open("closed_fist_le.pkl","wb") as f:
        pickle.dump(le, f)
    print("✅ Closed-fist refiner saved.")

def train_bw():
    X, y = load_images(BW_CLASSES)
    if len(X)==0:
        print("❌ No B/W data, skipping")
        return
    train_ds, val_ds, le = make_tf_datasets(X, y, binary=True)

    # clear bad weights
    cache = os.path.expanduser("~/.keras/models")
    fname = "mobilenet_v2_weights_tf_dim_ordering_tf_kernels_1.0_128_no_top.h5"
    fpath = os.path.join(cache, fname)
    if os.path.exists(fpath) and os.path.getsize(fpath)<5_000_000:
        os.remove(fpath)

    base = tf.keras.applications.MobileNetV2(
        input_shape=(*TARGET_SIZE,3),
        include_top=False, pooling='avg', weights='imagenet'
    )
    x   = tf.keras.layers.Dropout(0.3)(base.output)
    out = tf.keras.layers.Dense(1, activation='sigmoid')(x)
    model = tf.keras.Model(base.input, out)
    base.trainable = False

    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=10,
        callbacks=[
            tf.keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)
        ]
    )

    model.save("bw_refiner.keras")
    with open("bw_le.pkl","wb") as f:
        pickle.dump(le, f)
    print("✅ B vs W refiner saved.")

if __name__=="__main__":
    train_closed_fist()
    train_bw()
```

```python
# train_closed_fist_refiner.py

import os
import glob
import cv2
import numpy as np
import pickle
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# CONFIG
ROOT        = r"C:\Users\ishaa\Downloads\SpeakEz\Teacher\asl_alphabet_test"
IMG_EXTS    = ("*.jpg", "*.png")
TARGET_SIZE = (128, 128)
CLOSED_CLASSES = ["A", "E", "O", "S", "M", "N", "T"]

def load_images(classes):
    X, y = [], []
    for cls in classes:
        folder = os.path.join(ROOT, cls)
        if not os.path.isdir(folder):
            print(f"⚠️  Missing folder {folder}, skipping {cls}")
            continue
        for ext in IMG_EXTS:
            for fp in glob.glob(os.path.join(folder, ext)):
                img = cv2.imread(fp)
                if img is None:
                    continue
                img = cv2.resize(img, TARGET_SIZE)
                X.append(img)
                y.append(cls)
    X = np.array(X, dtype=np.float32) / 255.0
    y = np.array(y)
    print(f"📦 Loaded {len(X)} images for {classes}")
    return X, y

def make_tf_datasets(X, y, batch=32):
    le = LabelEncoder().fit(y)
    y_enc = le.transform(y)
    Xtr, Xvl, ytr, yvl = train_test_split(
        X, y_enc, test_size=0.2, random_state=42, stratify=y_enc
    )
    train_ds = tf.data.Dataset.from_tensor_slices((Xtr, ytr))
    train_ds = train_ds.shuffle(100).batch(batch).prefetch(1)
    val_ds = tf.data.Dataset.from_tensor_slices((Xvl, yvl))
    val_ds = val_ds.batch(batch).prefetch(1)
    return train_ds, val_ds, le

def train_closed_fist_refiner():
    X, y = load_images(CLOSED_CLASSES)
    if len(X) == 0:
        print("❌ No closed-fist data found, skipping training.")
        return

    train_ds, val_ds, le = make_tf_datasets(X, y)

    # Clear any truncated MobileNetV2 weights to force re-download
    cache = os.path.expanduser("~/.keras/models")
    fname = "mobilenet_v2_weights_tf_dim_ordering_tf_kernels_1.0_128_no_top.h5"
    fpath = os.path.join(cache, fname)
    if os.path.exists(fpath) and os.path.getsize(fpath) < 5_000_000:
        os.remove(fpath)

    # Build the model
    base = tf.keras.applications.MobileNetV2(
        input_shape=(*TARGET_SIZE, 3),
        include_top=False,
        pooling='avg',
        weights='imagenet'
    )
    x = tf.keras.layers.Dropout(0.3)(base.output)
    out = tf.keras.layers.Dense(len(CLOSED_CLASSES), activation='softmax')(x)
    model = tf.keras.Model(base.input, out)
    base.trainable = False

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    # Train
    model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=10,
        callbacks=[tf.keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)]
    )

    # Save model and label encoder
    model.save("closed_fist_refiner.keras")
    with open("closed_fist_le.pkl", "wb") as f:
        pickle.dump(le, f)
    print("✅ Closed-fist refiner saved.")

if __name__ == "__main__":
    train_closed_fist_refiner()
```
